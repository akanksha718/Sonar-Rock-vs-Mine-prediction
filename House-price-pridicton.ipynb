```python
X = df.drop(columns='medv', axis=1)
Y = df['medv']
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
xg.fit(X_train, Y_train)
```

```python
training_data_prediction = xg.predict(X_train)
print("R squared value :", metrics.r2_score(Y_train, training_data_prediction))
```

```python
print("Mean Absolute Error :", metrics.mean_absolute_error(Y_train, training_data_prediction))
```
